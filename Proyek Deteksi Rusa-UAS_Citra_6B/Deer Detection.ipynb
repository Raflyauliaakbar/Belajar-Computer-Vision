<h3>Hasil Confusion Matrix

Dari visualisasi confusion matrix di atas, kita bisa melihat bagaimana performa model dalam membedakan antara gambar yang berisi objek rusa (Deer) dan gambar yang hanya berisi latar belakang (background).

Model ini berhasil mengklasifikasikan 480 gambar rusa dengan benar, yang merupakan prediksi tepat terhadap objek yang memang ada â€” ini disebut True Positive. Hasil ini menunjukkan bahwa model cukup baik dalam mengenali objek targetnya.

Namun, model juga masih melakukan dua jenis kesalahan:

1. False Positive (FP)
Sebanyak 121 gambar latar belakang justru dikenali secara salah sebagai rusa. Ini seperti "alarm palsu", di mana model mengira ada rusa padahal sebenarnya tidak ada. Kesalahan ini bisa membuat model jadi kurang akurat saat memberikan prediksi positif.

2. False Negative (FN)
Kesalahan yang lebih besar terjadi ketika model gagal mendeteksi rusa yang sebenarnya ada. Sebanyak 266 gambar yang berisi rusa tidak terdeteksi, dan malah diklasifikasikan sebagai latar belakang. Ini mengindikasikan bahwa model melewatkan cukup banyak objek yang seharusnya bisa dikenali.

<h3>Analisis Akurasi Model

Jika dilihat dari metrik:

1. Precision (tingkat ketepatan) model adalah sekitar 79.87%. Artinya, dari semua gambar yang dikira rusa oleh model, sekitar 80%-nya memang benar-benar rusa.

2. Recall (tingkat kemampuan menangkap semua target) justru lebih rendah, yaitu hanya 64.34%. Ini menunjukkan bahwa model masih kurang optimal dalam mendeteksi semua rusa yang ada di dataset.

<h3>Kesimpulan

 Jadi model ini cukup baik dalam hal keakuratan saat menyatakan gambar tersebut mengandung rusa, tapi masih kesulitan dalam mendeteksi semua rusa yang ada. <h3>Hasil Presision Recall Curve<h1>

Gambar di atas menampilkan kurva Precision-Recall, yaitu grafik yang menggambarkan bagaimana performa model berubah tergantung pada ambang kepercayaan saat mendeteksi objek â€” dalam hal ini, rusa.

Apa Artinya?<br>
- Precision menunjukkan seberapa banyak dari prediksi "rusa" yang benar-benar rusa.<br>
- Recall menunjukkan seberapa banyak dari semua rusa yang berhasil ditemukan oleh model.

<h3>Bentuk Kurva

Pada awal grafik (bagian kiri bawah), saat model hanya memberikan prediksi yang paling yakin (recall masih rendah), precision sangat tinggi, bahkan hampir menyentuh 100%. Artinya, prediksi yang dibuat pada titik ini sangat akurat.

Namun, ketika model "dipaksa" untuk mendeteksi lebih banyak rusa (meningkatkan recall ke kanan), precision mulai turun. Ini menunjukkan bahwa model mulai membuat lebih banyak kesalahan, karena berusaha mendeteksi sebanyak mungkin target.

Penurunan precision ini menjadi tajam setelah recall lewat angka 0.6 â€“ 0.7, artinya saat model mulai mendeteksi lebih banyak objek, akurasinya menurun drastis.

<h3>Nilai mAP (mean Average Precision)

Nilai mAP@0.5 yang didapat adalah 0.662 atau 66.2%. Angka ini mencerminkan rata-rata performa model di semua titik dalam grafik. Semakin tinggi mAP, semakin baik model secara keseluruhan. Nilai 0.662 berarti model cukup bagus, tapi masih ada ruang untuk ditingkatkan.

<h3>Kesimpulan

1. Kurva ini memperkuat hasil dari confusion matrix sebelumnya.

2. Precision tinggi di awal menunjukkan model bisa sangat akurat saat yakin.

3. Recall yang lebih rendah membuktikan bahwa model masih melewatkan cukup banyak target.

4. Ada kompromi antara akurasi dan kelengkapan deteksi â€” semakin banyak objek yang coba dideteksi, semakin besar kemungkinan salah prediksi.

Grafik ini memvisualisasikan bagaimana performa model berubah selama 100 epoch pelatihan.

Kurva Loss (Kerugian): Anda akan melihat semua kurva loss (misalnya box_loss, cls_loss) terus menurun baik untuk data train maupun val (validasi). Tidak Overfitting: Fakta bahwa kurva validasi (val) mengikuti tren kurva training (train) dengan ketat adalah pertanda sangat baik. Ini menunjukkan model tidak overfitting (tidak hanya menghafal data training) dan mampu menggeneralisasi pengetahuannya dengan baik untuk data baru yang belum pernah dilihat.

**<h4>Kesimpulan Hasil Training Antara Epoch 50 dan Epoch 100<h4>**

<h4>Penjelasan Hasil<h4>

1. Peningkatan Performa: Pelatihan hingga 100 epoch berhasil meningkatkan metrik utama mAP50-95 dari 0.77 menjadi 0.811. Ini menunjukkan bahwa model menjadi lebih akurat dan andal dalam mendeteksi objek dengan berbagai tingkat presisi. Peningkatan ini, meskipun tidak sebesar lompatan performa di awal-awal epoch, tetap signifikan dan membuktikan bahwa pelatihan tambahan memberikan manfaat.

2. Mencapai Titik Jenuh (Diminishing Returns): Grafik results.png dari kedua pelatihan menunjukkan bahwa sebagian besar "pembelajaran" terjadi pada 50 epoch pertama. Setelah itu, kurva loss semakin landai dan kurva mAP naik lebih lambat. Ini adalah fenomena wajar yang disebut diminishing returns, di mana setiap epoch tambahan memberikan peningkatan yang semakin kecil.

3. Tidak Ada Tanda Overfitting: Salah satu kekhawatiran dari pelatihan yang terlalu lama adalah overfitting (model terlalu "hafal" data training). Namun, berdasarkan grafik, kurva validasi (val) masih mengikuti kurva training (train) dengan baik. Ini berarti model dengan 100 epoch masih mampu melakukan generalisasi dengan baik pada data baru dan tidak mengalami overfitting yang signifikan.
<br>

Pelatihan model dengan 100 epoch menghasilkan performa yang secara definitif lebih baik dibandingkan dengan 50 epoch. Peningkatan metrik mAP (terutama mAP50-95) membuktikan bahwa model menjadi lebih akurat dan dapat diandalkan.

Meskipun peningkatan setelah 50 epoch tidak lagi drastis, hasil akhir menunjukkan bahwa investasi waktu untuk pelatihan tambahan hingga 100 epoch sepadan dengan peningkatan kualitas deteksi. Oleh karena itu, untuk kasus dataset ini, model yang dilatih selama 100 epoch adalah pilihan terbaik karena memberikan akurasi tertinggi tanpa menunjukkan tanda-tanda overfitting yang merugikan.
